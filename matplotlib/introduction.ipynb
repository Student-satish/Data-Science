```python
years = [1999, 2002, 2003, 2005, 2008, 2011]
sachin_runs = [1328, 1161, 1141, 1231, 1006, 513]
sehwag_runs = [270, 659, 1001, 623, 1157, 491]
# format strings color -> marker -> line = 'ro--' example 
plt.plot(years,sachin_runs,'ro--',label="sachin");
plt.plot(years,sehwag_runs,'g^:',label="sehwag");
plt.xlabel("years");
plt.ylabel("runs scored");
plt.title("performance comparision");
plt.legend();
```

```python
# color and linestyle arguments
years = [1999, 2002, 2003, 2005, 2008, 2011]
sachin_runs = [1328, 1161, 1141, 1231, 1006, 513]
sehwag_runs = [270, 659, 1001, 623, 1157, 491]
plt.plot(years,sachin_runs,color="orange",linestyle="--",linewidth=3,label="sachin");
plt.plot(years,sehwag_runs,color="red",linestyle="-.",linewidth=3,label="sehwag");
plt.xlabel("years");
plt.ylabel("Runs Scored");
plt.title("performance comparision");
plt.tight_layout();
plt.grid(True);
plt.legend();
plt.style.use('petroff10',);
plt.show();
```
