```python
import matplotlib.pyplot as plt;
```

```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
```

```python
plt.bar(years,runs);
plt.xlabel("years");
plt.ylabel("runs");
plt.title("dhoni yearly runs");
plt.tight_layout();
plt.show();
```

```python
import numpy as np
```

```python
x = np.arange(len(years));
```

```python
x
```

```python
sachin = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
sehwag = [0, 200, 900, 1400, 1600, 1800, 1500, 1100, 800, 0]
kohli  = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
```

```python
 width = 0.25
```

```python
plt.bar(x-width,sachin,width=width,label="sachin");
plt.bar(x,sehwag,width=width,label="sehwag");
plt.bar(x+width,kohli,width=width,label="kohli");
plt.xlabel("years");
plt.ylabel("runs");
plt.title("runComparision");
plt.xticks(x,years);
plt.legend();
plt.tight_layout();
plt.show();
```

```python
# horizontal bar plots
players = ["kohli","dhoni","sachin","raina"];
runs = [1800,1500,1000,1200];
plt.barh(players,runs);
plt.xlabel("runs");
plt.ylabel("player");
plt.title("runs comparision");
plt.tight_layout();
plt.show();
```

```python
players = ["kohli","dhoni","sachin","raina"];
runs = [1800,1500,1000,1200];
plt.bar(players,runs);
for i in range(len(players)):
    plt.text(i,runs[i]+20,str(runs[i]),ha="center");
plt.xlabel("player");
plt.ylabel("runs");
plt.title("runs comparision");
plt.tight_layout();
plt.show();
```
