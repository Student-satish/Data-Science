```python
import matplotlib.pyplot as plt
```

```python
labels = ["kohli","sachin","sehwag","dhoni"];
runs = [1000,1100,800,700];
```

```python
plt.style.use("ggplot");
```

```python

colors = ["red","blue","yellow","green"];
# matplotlib wedge
plt.pie(runs,labels=labels,colors=colors,wedgeprops = {'edgecolor':'black','linewidth':3,'linestyle':'-.','fill':True});
plt.title("Runs of indian Batsmen");
plt.show();
```

```python
explode = [0.1,0,0,0];
plt.pie(runs,labels=labels,colors=colors,shadow=True,explode=explode,startangle=100,autopct='%1.1f%%');
plt.title("Runs of indian Batsmen");
plt.show();
```

```python
import matplotlib.pyplot as plt
```

```python
days = [1,2,3,4,5,6,7];
studying = [3,4,3,5,4,3,4];
playing = [2,2,1,1,2,3,2];
watching_tv = [2,1,2,2,1,1,1];
sleeping = [5,5,6,5,6,5,5];
labels = ['studying','playing','watching_tv','sleeping'];
colors = ['red','yellow','orange','green'];
```

```python
plt.figure(figsize=(10,6));
plt.stackplot(days,studying,playing,watching_tv,sleeping,labels = labels,colors = colors,alpha=0.4);
plt.grid(True);
plt.legend(loc='upper left');
plt.xlabel("Days");
plt.ylabel("Hours");
plt.title("Student Time Distribution");
plt.show();
```
